```python
import pandas as pd
from datetime import datetime
```

```python
# Set up dataframe
index = ['dedup','blackscholes','ferret','freqmine','canneal','fft']

measurements = pd.DataFrame(columns=['none', 'cpu', 'l1d', 'l1i', 'l2', 'llc', 'membw'], index=index)
measurements
```

```python
# Load measurements
# ()

#                                     none     	cpu	         l1d	     l1i	      l2	     llc	    membw
measurements.loc['dedup'] =        ['1m0.930s','1m33.168s','1m15.124s','2m30.307s','1m18.039s','2m16.028s','1m42.312s']
measurements.loc['blackscholes'] = ['4m14.322s','5m58.046s','5m48.282s','8m0.563s','5m42.328s','7m15.510s','6m7.423s']
measurements.loc['ferret'] =       ['0m13.189s','0m25.059s','0m13.222s','0m35.183s','0m13.326s','0m35.108s','0m27.702s']
measurements.loc['freqmine'] =     ['0m6.569s','0m12.716s','0m6.500s','0m12.876s','0m6.456s','0m11.986s','0m10.197s']
measurements.loc['canneal'] =      ['0m14.475s','0m18.226s','0m19.178s','0m22.950s','0m18.676s','0m29.911s','0m21.051s']
measurements.loc['fft'] =          ['0m8.317s','0m11.004s','0m10.648s','0m15.865s','0m10.648s','0m16.757s','0m12.740s']
measurements
```

```python
# convert from timestr to seconds

# Convert timestring returned by time utility to a float of seconds
def tstr_to_sec(time_string):
    nofrag, frag = time_string.split('.')
    pt = datetime.strptime(nofrag,'%Mm%S')   
    total_seconds = pt.second + pt.minute*60 + float('.' + frag[:-1])
    return total_seconds

measurements_s = measurements.applymap(tstr_to_sec)
measurements_s
```

```python
# Normalize rows
def normalize_speedup(values):
    return [ round(val/values['none'], 2)  for val in values]

measurements_sn = measurements_s.apply(normalized_speedup, axis=1)

measurements_sn
```
